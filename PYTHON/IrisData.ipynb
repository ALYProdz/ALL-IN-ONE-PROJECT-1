## Machine Learning 

ML - Model

```python
x = pd.DataFrame(iris.data, columns = iris.feature_names)
```

```python
X = df[["sepal length (cm)","sepal width (cm)","petal length (cm)","petal width (cm)"]]
```

```python
y = pd.DataFrame(iris.target)
```

```python
y
```

```python
#Select the model
model = LinearRegression()
```

```python
#Fit the model
model.fit(X,y)
```

```python
#Train, test, and split the model
x_train, x_test,y_train, y_test = train_test_split(X,y, test_size = 0.33, random_state = 44) 
```

```python
x_train
```

```python
x_test
```

```python

#Visualize
```

```python
plt.subplot(1,1,1)
sns.regplot(data = df, x ="sepal length (cm)",y = "sepal width (cm)", color = 'blue')
sns.regplot(data = x_train, x ="sepal length (cm)",y = "sepal width (cm)", color = "red")
plt.title("Visualization ML & Normal Df")
```

```python
plt.subplot(1,1,1)
sns.regplot(data = df, x ="sepal length (cm)",y = "sepal width (cm)", color = 'blue')
sns.regplot(data = x_test, x ="sepal length (cm)",y = "sepal width (cm)", color = "red")
plt.title("Visualization ML tst & Normal Df")
```

```python
#Fit x_train & y_train
model.fit(x_train, y_train)
```

```python
#Find model coef
model.coef_
```

```python
#Predict x_train
x_tr_pr = model.predict(x_train)
x_tr_pr
```

```python
#Predict the x_test
x_test_predict = model.predict(x_test)
```

```python
x_test_predict
```

```python
#Find the mean of the prediction test
#tes_pre_avg = np.mean(np.power(x_test_predict - x_test),2)np

tes_pre_avg = np.mean(np.power((x_test_predict - x_test),2))
```
